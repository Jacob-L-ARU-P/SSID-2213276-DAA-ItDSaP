## Displaying the feature names, data types and size of a data set
In the previous task you looked at a table of data. It contains data for every day of either 1987 or 2015 from May to October. It is composed of columns and rows.

* Each column of the data table represents one **feature**. Each feature holds a different data value. Note that the term *field* is often used instead of *feature*.
* Each row of the table represents one **record**. A record holds all the data values for a single date at a single weather station.

The next block of code will output information about the data set.

> Run the code now and check the result

```python
# explore the data types
weather_data.info()
```

The code outputs a list of the features in this table. It also shows the data type of each feature. The main data types you will encounter are:
* `int64`: integers (64 *bits* are used to store the number)
* `float64`: a *floating point* number: i.e. includes a decimal point (64 *bits* are used to store the number)
* `object`: this includes text features

It is important to know how the features are stored as this has an impact on the type of analysis you can do, for example you will not be able to calculate a mean of a set of text values.

# Extension: Pre-processing the data
## Cleaning the Daily Total Rainfall values
Sometimes a data set has missing data values. In other cases the values might be obviously incorrect. For example in a list of temperatures you might see a single value higher than 1000. This is obviously wrong and will affect your summary values for that data set.

Finding and fixing these problems is called *pre-processing* (or *cleaning*) the data. In this part of the activity you will find and fix some problems with the data set.

> Run the code below to output the summary values for `Daily Total Rainfall`

```python
# calculate the grouped summary statistics
weather_data['Daily Total Rainfall'].describe()
```

* How are the results different from the results for 'Daily Mean Temperature'? 
* Look back at the output for `info()` at the top of this notebook. How is the rainfall data different?

Many of the values you might expect to see (such as 'mean') are not calculated for Daily Total Rainfall. That is because the feature has been imported as an `object` or text feature. When importing if any of the values are text (other than "N/A") then the whole feature is imported as text. The computer can't create summary statistics from a mixture of data types. So you need to process the data before you can see summary values. 

The code in box below cleans the data. The value of "tr" represents a trace amount of rain less than 0.05mm therefore we are replacing it with a value in the middle of the range 0-0.05 and converting it to a number.

The first command creates a copy of the feature as a new feature called 'Rainfall'.

The second command replaces â€œtrâ€ with 0.025 and changes every value from "object" to â€œfloatâ€ data type so numerical statistics can be calculated.

> Run the code to clean the data.

```python
# create a new column called Rainfall which is a copy of the Daily Total Rainfall column
weather_data['Rainfall'] = weather_data['Daily Total Rainfall']

# replace any instances of 'tr' with 0.025 and change the type to float
weather_data['Rainfall'] = weather_data['Rainfall'].replace({'tr': 0.025}).astype('float')

# desplay the data to check
weather_data
```
